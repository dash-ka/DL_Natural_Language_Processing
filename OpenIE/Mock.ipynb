```python
s = "Accordingly , the 1962 Roman Missal , the edition whose continued use as an extraordinary form of the Roman Rite is authorized by the motu proprio `` Summorum Pontificum '' , also has no mention of her ."
nlp_sentence = nlp(s)

g = generate_graph(nlp_sentence, fold_preps=True, collapse_of=True, skip_tokens = ["punct"])
fig = plt.figure(figsize=(22, 20))
draw_graph(g)

triples_dict = triples(g)
for verb, subj_obj_list in triples_dict.items():
    for subj, obj, suffix in subj_obj_list:
        print("< " + g.nodes[subj[0]]["node_text"] + " >",                      # formatting <subj>
              g.nodes[verb]["node_text"] if isinstance(verb, int) else verb ,   # formatting <predicate>
              "[" + suffix + "]",                                               # formatting suffix
              "< " + " ".join([g.nodes[o]["node_text"] for o in obj]) + " >" )     
```
